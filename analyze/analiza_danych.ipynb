```python
data = pd.read_csv("ready_dataset_with_nulls_as_incorrect_data.csv")

# Wykres zaleÅ¼noÅ›ci MMSE i FunctionalAssessment z podziaÅ‚em na obecnoÅ›Ä‡ urazÃ³w gÅ‚owy

# Tworzenie ukÅ‚adu z dwoma wykresami
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Wykres 1: MMSE vs FunctionalAssessment z podziaÅ‚em na uraz gÅ‚owy
scatter1 = sns.scatterplot(
    x=data["MMSE"], y=data["FunctionalAssessment"], 
    hue=data["HeadInjury"], palette={0: 'blue', 1: 'red'}, ax=axes[0]
)
axes[0].set_xlabel("MMSE (Mini-Mental State Examination)")
axes[0].set_ylabel("FunctionalAssessment (Ocena funkcji)")
axes[0].set_title("WpÅ‚yw urazu gÅ‚owy na funkcje poznawcze i oceny funkcjonalne")

handles, labels = scatter1.get_legend_handles_labels()
axes[0].legend(handles=handles, labels=["Nie", "Tak"], title="Head Injury (Uraz gÅ‚owy)")

# Cholesterol LDL vs CiÅ›nienie skurczowe a spoÅ¼ycie alkoholu
scatter2 = axes[1].scatter(
    data["CholesterolLDL"], data["SystolicBP"], 
    c=data["AlcoholConsumption"], cmap="viridis", edgecolors='k'
)
axes[1].set_xlabel("Cholesterol LDL (mg/dL)")
axes[1].set_ylabel("CiÅ›nienie skurczowe (mmHg)")
axes[1].set_title("WpÅ‚yw spoÅ¼ycia alkoholu na cholesterol LDL i ciÅ›nienie skurczowe")

cbar = fig.colorbar(scatter2, ax=axes[1])
cbar.set_label("SpoÅ¼ycie alkoholu (jednostki/tydzieÅ„)")

plt.tight_layout()
plt.savefig("AlcoholConsumption_on_HeadInjury_CholesterolLDL")
plt.show()
```

```python
data = pd.read_csv("ready_dataset_with_nulls_as_incorrect_data.csv")

# PorÃ³wnanie wynikÃ³w MMSE i FunctionalAssessment w rÃ³Å¼nych grupach wiekowych

bins = [60, 65, 70, 75, 80, 85, 90, 95]
labels = ["60-64", "65-69", "70-74", "75-79", "80-84", "85-89", "90+"]
data["AgeGroup"] = pd.cut(data["Age"], bins=bins, labels=labels, include_lowest=True)

sns.set_style("whitegrid")
fig, axes = plt.subplots(1, 2, figsize=(14, 6))  # 1 wiersz, 2 kolumny

# MMSE
sns.boxplot(x="AgeGroup", y="MMSE", data=data, ax=axes[0], width=0.5)
axes[0].set_xlabel("Grupa wiekowa")
axes[0].set_ylabel("Wynik MMSE")
axes[0].set_title("PorÃ³wnanie wynikÃ³w MMSE w rÃ³Å¼nych grupach wiekowych")

# FunctionalAssessment
sns.boxplot(x="AgeGroup", y="FunctionalAssessment", data=data, ax=axes[1], width=0.5)
axes[1].set_xlabel("Grupa wiekowa")
axes[1].set_ylabel("Wynik FunctionalAssessment")
axes[1].set_title("PorÃ³wnanie wynikÃ³w FunctionalAssessment w rÃ³Å¼nych grupach wiekowych")

# WyÅ›wietlenie wykresÃ³w
plt.tight_layout()  # Optymalne rozmieszczenie wykresÃ³w
plt.savefig("FunctionalAssessment_MMSE_Age")
plt.show()
```

```python
data = pd.read_csv("ready_dataset_with_nulls_as_incorrect_data.csv")

# Wykres zaleÅ¼noÅ›ci MMSE i FunctionalAssessment z podziaÅ‚em na obecnoÅ›Ä‡ choroby Alzheimera
plt.figure(figsize=(8,6))
scatter = sns.scatterplot(x=data["MMSE"], y=data["FunctionalAssessment"], hue=data["Diagnosis"], palette={0: 'blue', 1: 'green'})
plt.xlabel("MMSE (Mini-Mental State Examination)")
plt.ylabel("FunctionalAssessment (Ocena funkcji)")
plt.title("WpÅ‚yw obecnoÅ›ci choroby Alzheimera na funkcje poznawcze i oceny funkcjonalne")

# Ustawienie legendy
handles, labels = scatter.get_legend_handles_labels()
scatter.legend(handles=handles, labels=["Brak choroby", "Choroba Alzheimera"], title="Diagnosis (Diagnoza)")
plt.savefig("Alzheimer_FunctionalAssessment_MMSE")
plt.show()
```

```python
data = pd.read_csv("ready_dataset_with_nulls_as_incorrect_data.csv")

data["Diagnosis"] = data["Diagnosis"].map({0: "Brak choroby", 1: "Choroba Alzheimera"})

bins = [60, 65, 70, 75, 80, 85, 90, 95]  # PrzedziaÅ‚y wieku
labels = ["60-64", "65-69", "70-74", "75-79", "80-84", "85-89", "90-94"]
data["AgeGroup"] = pd.cut(data["Age"], bins=bins, labels=labels, include_lowest=True)

palette = {"Brak choroby": "blue", "Choroba Alzheimera": "green"}

# PorÃ³wnanie wynikÃ³w MMSE w rÃ³Å¼nych grupach wiekowych z podziaÅ‚em na chorobÄ™ Alzheimera
plt.figure(figsize=(10,6))
sns.boxplot(x="AgeGroup", y="MMSE", data=data, hue="Diagnosis", palette=palette)
plt.xlabel("Grupa wiekowa")
plt.ylabel("Wynik MMSE")
plt.title("PorÃ³wnanie wynikÃ³w MMSE w zaleÅ¼noÅ›ci od wieku i obecnoÅ›ci choroby Alzheimera")
plt.legend(title="Diagnoza")
plt.savefig("MMSE_Alzheimer_age")
plt.show()

# PorÃ³wnanie wynikÃ³w FunctionalAssessment w rÃ³Å¼nych grupach wiekowych z podziaÅ‚em na chorobÄ™ Alzheimera
plt.figure(figsize=(10,6))
sns.boxplot(x="AgeGroup", y="FunctionalAssessment", data=data, hue="Diagnosis", palette=palette)
plt.xlabel("Grupa wiekowa")
plt.ylabel("Wynik FunctionalAssessment")
plt.title("PorÃ³wnanie wynikÃ³w FunctionalAssessment w zaleÅ¼noÅ›ci od wieku i obecnoÅ›ci choroby Alzheimera")
plt.legend(title="Diagnoza")
plt.savefig("FunctionalAssessment_Alzheimer_age")
plt.show()
```
